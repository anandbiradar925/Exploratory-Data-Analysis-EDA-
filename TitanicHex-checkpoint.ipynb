```python
df.rename(columns={'Passenger_id': 'PASSENGER_ID'}, inplace=True)
df.rename(columns={'Age': 'AGE'}, inplace=True)
df.rename(columns={'Fare': 'FARE'}, inplace=True)
df.rename(columns={'Sex': 'SEX'}, inplace=True)
```

```python
df.rename(columns={'sibsp': 'SIBSP'}, inplace=True)
df.rename(columns={'Parch': 'PARCH'}, inplace=True)
df.rename(columns={'Pclass': 'P-CLASS'}, inplace=True)
df.rename(columns={'Embarked': 'EMBARKED'}, inplace=True)
df.rename(columns={'survived': 'SURVIVED'}, inplace=True)
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
Q1 = df['AGE'].quantile(0.25)
Q3 = df['AGE'].quantile(0.75)
IQR = Q3 - Q1

outliers_age = df[(df['AGE'] < Q1 - 1.5*IQR) | 
                  (df['AGE'] > Q3 + 1.5*IQR)]

outliers_age
```

```python
outliers_age.shape[0]
df_clean = df[
    (df['AGE'] >= Q1 - 1.5*IQR) & 
    (df['AGE'] <= Q3 + 1.5*IQR)]
df_clean.shape
df = df[df['FARE'] > 0]
```

```python
df.groupby('P-CLASS')['SURVIVED'].mean()
df.groupby('SEX')['SURVIVED'].mean()
```

```python
sex_counts = df['SEX'].value_counts()

labels = ['Male', 'Female']

plt.figure(figsize=(6,6))
plt.pie(
    sex_counts,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Gender Distribution')
plt.show()



counts = df['SURVIVED'].value_counts()

labels = ['Unsurvived', 'Survived']

plt.figure(figsize=(6,6))
plt.pie(
    counts,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Survival Distribution')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
male_counts = df[df['SEX'] == 0]['SURVIVED'].value_counts()
axes[0].pie(
    male_counts,
    labels=male_counts.index.map({0: 'Unsurvived', 1: 'Survived'}),
    autopct='%1.1f%%',
    startangle=90
)
axes[0].set_title('Male Survival Rate')
female_counts = df[df['SEX'] == 1]['SURVIVED'].value_counts()
axes[1].pie(
    female_counts,
    labels=female_counts.index.map({0: 'Unsurvived', 1: 'Survived'}),
    autopct='%1.1f%%',
    startangle=90
)
axes[1].set_title('Female Survival Rate')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df[df['SURVIVED'] == 0]['AGE'], bins=20, alpha=0.6, label='Unsurvived')
plt.hist(df[df['SURVIVED'] == 1]['AGE'], bins=20, alpha=0.6, label='Survived')
plt.xlabel('AGE')
plt.ylabel('Count')
plt.title('Age Distribution by Survival')
plt.legend()
plt.show()
```
